```python
import csv
```

```python
correspondenceFileJupyter = 'output/correspondences.csv'
correspondenceFileSparql = 'output/correspondences-sparql.csv'
```

```python
correspondencesJupyter = {}
with open(correspondenceFileJupyter, 'r') as f:
    reader = csv.DictReader(f)
    for row in reader:
        correspondencesJupyter[row['uuid']] = row['filename']
```

```python
correspondencesSparql = {}
with open(correspondenceFileSparql, 'r') as f:
    reader = csv.DictReader(f)
    for row in reader:
        correspondencesSparql[row['uuid']] = row['filename']
    
```

```python
for key,value in correspondencesJupyter.items():
    if key not in correspondencesSparql and value:
        print(f'Key {key} not present in SPARQL (value {value})')
    elif key in correspondencesSparql and correspondencesSparql[key] != value:
        print(f'Key {key} is different in SPARQL: {correspondencesSparql[key]} instead of {value})')
```
